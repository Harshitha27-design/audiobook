```python
!pip install transformers datasets accelerate gradio gTTS

import gradio as gr
from transformers import pipeline
from gtts import gTTS
import tempfile
import os

# Hugging Face model for text rewriting (tone control)
rewriter = pipeline("text2text-generation", model="google/flan-t5-large")

# Function for tone rewriting
def rewrite_text(text, tone):
    prompt = f"Rewrite the following text in a {tone} tone, while keeping the meaning the same:\n\n{text}"
    result = rewriter(prompt, max_length=300, temperature=0.7)[0]['generated_text']
    return result

# Function for text-to-speech
def text_to_speech(original_text, tone, voice):
    rewritten = rewrite_text(original_text, tone)
    # Convert to audio using gTTS (Google TTS, simple & fast)
    tts = gTTS(rewritten, lang="en", tld="com", slow=False)
    tmp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
    tts.save(tmp_file.name)
    return rewritten, tmp_file.name

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("# 🎧 EchoVerse - AI Audiobook Generator (Hugging Face Version)")
    gr.Markdown("Upload text or type below, choose a tone & voice, and get expressive audio output!")

    with gr.Row():
        text_input = gr.Textbox(label="Enter Text", lines=8, placeholder="Paste your text here...")

    tone = gr.Radio(["Neutral", "Suspenseful", "Inspiring"], label="Select Tone", value="Neutral")
    voice = gr.Dropdown(["Lisa", "Michael", "Allison", "Kate"], label="Select Voice", value="Lisa")

    btn = gr.Button("Generate Audiobook 🎙️")

    rewritten_text = gr.Textbox(label="Rewritten Text", lines=8)
    audio_output = gr.Audio(label="Generated Audio", type="filepath")

    btn.click(text_to_speech, inputs=[text_input, tone, voice], outputs=[rewritten_text, audio_output])

demo.launch()
```
